```python
cols_to_drop = ['Poster_Link','Runtime','Gross']
df.drop(cols_to_drop,axis=1,inplace=True)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
df['text']= df['Series_Title']+' '+df['Genre']+' '+df['Overview']+' '+df['Director']+' '+df['Star1']+' '+df['Star2']+' '+df['Star3']+' '+df['Star4']
```

```python
vec = TfidfVectorizer(stop_words='english',ngram_range=(1,3))
tfidf_mat = vec.fit_transform(df['text']).toarray()
print(tfidf_mat.shape)
```

```python
sim = cosine_similarity(tfidf_mat, tfidf_mat).round(2)
```

```python
def get_index_from_name(name):
    return df[df['Series_Title']==name].index[0]
```

```python
 idx = get_index_from_name('The Godfather')
print(df.iloc[idx])
```

```python
def recommend_movie(movie_name,k=5):
    idx = get_index_from_name(movie_name)   # get index of users movie
    sim_scores = list(enumerate(sim[idx]))  # get similarity scores for all movies
    sim_scores.sort(key=lambda x: x[1],reverse=True)    # Sort the list so that the heighest scores
    movie_indices = [i[0] for i in sim_scores]      # Get movie indices
    return df.iloc[movie_indices]['Series_Title'].head(k)
```

```python
recommend_movie('The Godfather',10)
```
