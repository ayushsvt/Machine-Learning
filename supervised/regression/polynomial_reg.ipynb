```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly_express as px

from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score 
```

```python
url=('https://raw.githubusercontent.com/digipodium/Datasets/main/Position_Salaries.csv')
df=pd.read_csv(url)
df.head()
```

```python
X = df[['Level']].values
y = df['Salary'].values
```

```python
reg1 = LinearRegression()
reg1.fit(X,y)
ypred = reg1.predict(X)
df['y_pred'] = ypred
ax = df.plot(kind='scatter',x='Level',y='Salary',figsize=(10,5),color='red',title='Simple Linear Regression',label='original')
df.plot(kind='line',x='Level',y='y_pred',ax=ax)
plt.show()
print("MAE",mean_absolute_error(y,ypred))
print("MSE",mean_squared_error(y,ypred))
print("R2",r2_score(y,ypred))

```

```python
pf= PolynomialFeatures(degree=4)
X_poly= pf.fit_transform(X)

reg1 = LinearRegression()
reg1.fit(X_poly,y)
ypred = reg1.predict(X_poly)
df['y_pred_4'] = ypred
ax = df.plot(kind='scatter',x='Level',y='Salary',figsize=(10,5),color='red',title='Simple Linear Regression Degree 5',label='original')
df.plot(kind='line',x='Level',y='y_pred_4',ax=ax)
plt.show()
print("MAE",mean_absolute_error(y,ypred))
print("MSE",mean_squared_error(y,ypred))
print("R2",r2_score(y,ypred))

```
