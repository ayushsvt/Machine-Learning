```python
def check_weight(w):
    if w < 2000:
        return 'light'
    elif w < 3100:
        return 'medium'
    else:
        return 'heavy'
```

```python
df['curb-weight'].apply(check_weight)
```

```python
df['class'] = df['curb-weight'].apply(check_weight)
```
